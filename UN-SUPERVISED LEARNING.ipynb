## CLUSTERING MODELS

### K-MEAN CLUSTERING

```python
from sklearn.cluster import KMeans
```

```python
model_kmeans=KMeans(n_clusters=3)
model_kmeans.fit(data1)
km_pred=model_kmeans.predict(data1)
```

```python
km_pred
```

### ELBOW METHOD
###### USED TO OBTAIN THE OPTIMAL NO OF CLUSTERS

```python
model_kmeans.inertia_
```

```python
inertia=[]
for k in range(1,10):
    model=KMeans(n_clusters=k)
    model.fit(data1)
    inertia.append(model.inertia_)
```

```python
plt.plot(range(1,10),inertia,'-o')
plt.title('Elbow method')
plt.xlabel('no of clusters')
plt.ylabel('inertia')
plt.show()
```

# PCA

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(data1)
```

```python
scaled_data=pd.DataFrame(scaled_data,columns=data1.columns)
```

```python
scaled_data.head()
```

```python
# DOING PCA
from sklearn.decomposition import PCA
pca = PCA(n_components =5)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
x_pca.shape
```

```python
# variance determination for knowing the information loss 
# if variance is high information loss is less
pca.explained_variance_ratio_
```

```python
principalComponents = pca.fit_transform(scaled_data)
# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_, color='green')
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)
# Save components to a DataFrame
PCA_components = pd.DataFrame(principalComponents)
```

```python
#plot the graph to find the principal components
plt.figure(figsize=(10,10))
plt.plot(np.cumsum(pca.explained_variance_ratio_), 'ro-')
plt.grid()
```

```python
#AS per the graph, we can see that 4 principal components attribute for 90% of variation in the data. 
#we shall pick the first 4 components for our prediction.
new_pca = PCA(n_components=4)
new_x = new_pca.fit_transform(scaled_data)
new_x
```
